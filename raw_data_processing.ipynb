```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
solar_df = pd.read_csv('data/Solar.csv')
print(solar_df.shape)
solar_df.head(10)
```

```python
hh_df = pd.read_csv('data/data_ws.csv')
print(hh_df.shape)
hh_df.head(15)
```

```python
hh_df = hh_df[['dataid', 'localhour', 'use', 'temperature', 'cloud_cover','wind_speed']]
hh_df.head()
```

```python
unique_dataid = np.unique(hh_df['dataid'])
hhids = list(unique_dataid)
print(hhids)
```

```python
df_by_hh = {}
for hhid in hhids:
    df_by_hh[hhid] = []
```

```python
solar_df.columns = ['date', 'localhour', 'GH']
solar_df.head()
```

```python
time_gh = {}
```

```python
for index, row in solar_df.iterrows():
    date = row.date.split('/')
    if len(date[0]) < 2:
        date[0] = '0' + date[0]
    if len(date[1]) < 2:
        date[1] = '0' + date[1]
    time_str = '{}-{}-{} {}'.format(date[-1], date[0], date[1], row.localhour)
    time_gh[time_str] = row.GH
```

```python
len(time_gh)
```

```python
hh_df.localhour[:10]
```

```python
hh_df['GH'] = 0.0
hh_df['is_weekday'] = 0
hh_df.iloc[-50000]
```

```python
used = ['localhour', 'use', 'temperature', 'cloud_cover', 'wind_speed', 'GH', 'is_weekday']
```

```python
for index, row in hh_df.iterrows():
    try:
        row.GH = time_gh[row.localhour[:-6]]
        if row.localhour[-1] == '5':
            row.is_weekday = int(datetime.strptime(str(row.localhour), "%Y-%m-%d %H:%M:%S-05").weekday() < 5)
        else:
            row.is_weekday = int(datetime.strptime(str(row.localhour), "%Y-%m-%d %H:%M:%S-06").weekday() < 5)
        df_by_hh[row.dataid].append(row[used])
    except:
        print(row.localhour[:-6])
```

```python
for i in hhids:
    print('\nProcessing data of household {}, {} samples in total.'.format(i, len(df_by_hh[i])))
    df = pd.DataFrame(data=df_by_hh[i], columns=used)
    print(df.head())
    df.to_csv('data/processed_hhdata_{}.csv'.format(i))
```
